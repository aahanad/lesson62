```python
import pandas as pd
data=pd.DataFrame({
    "Name":["sophie","lizzy","annabelle"],
    "Age":[12,13,11],
    "City":["Redmond","Kirkland","Bellevue"]
})
```

```python
#Basic utilities to explore data:
#head– shows the first few rows
# describe() – gives summary statistics
# info() – shows data types and missing values
# shape – shows the number of rows and columns
# columns – lists all column names
print(data.head())
print(data.describe())
print(data.info())
print(data.shape)
print(data.columns)
```

```python
#adding a new column
data["Grades"]=[98,73,62]
data
```

```python
#adding a new row
new_row=pd.DataFrame({"Name":"emmy","Age":14,"City":"Seattle","Grades":99

},index=[0])
data=pd.concat([new_row,data]).reset_index(drop=True)
data
```
