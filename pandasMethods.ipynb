```python
import pandas as pd
data=pd.DataFrame({
    "Name":["sophie","lizzy","annabelle"],
    "Age":[12,13,11],
    "City":["Redmond","Kirkland","Bellevue"]
})
```

```python
#Basic utilities to explore data:
#head– shows the first few rows
# describe() – gives summary statistics
# info() – shows data types and missing values
# shape – shows the number of rows and columns
# columns – lists all column names
print(data.head())
print(data.describe())
print(data.info())
print(data.shape)
print(data.columns)
```

```python
data["City"]
```

```python
data["Age"].max()
```

```python
#adding a new column
data["Grades"]=[98,73,62]
data
```

```python
#adding a new row
new_row=pd.DataFrame({"Name":"emmy","Age":14,"City":"Seattle","Grades":99

},index=[0])
data=pd.concat([new_row,data]).reset_index(drop=True)
data
```

```python

#deleting a row
data=data[data["Name"]!="emmy"]
data
```

```python
#deleting a column
data.drop(["City"],axis=1,inplace=True)
data
```

```python
#selecting and storing one or more columns

#.loc[] (Label-based indexing)
#.iloc[] (Integer-based indexing)
d2=data.iloc[0:2,0:2]
d2
```

```python
data.set_index("Name",inplace=True)
first=data.loc["annabelle"]
first
```
